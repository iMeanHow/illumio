```python
import pandas as pd
import numpy as np
import os
import random as r
```

```python
rules=[['' for i in range(4)] for j in range(5000)]
direction=['inbound','outbound']
port=['tcp','udp']
for i in range(5000):
    rules[i][0]=r.choice(direction)
    rules[i][1]=r.choice(port)
    port_flag=r.randint(1,2)  # 1 just single port 2 a range of port
    if(port_flag==1):
        rules[i][2]=str(r.randint(1,65535))
    else:
        p1=r.randint(1,65534)
        p2=r.randint(1,65535-p1)
        rules[i][2]=str(p1)+'-'+str(p1+p2)
    port_flag=r.randint(1,2)  # 1 just single ip 2 a range of ip
    if(port_flag==1):
        rules[i][3]=str(r.randint(1,255))+'.'+str(r.randint(1,255))+'.'+str(r.randint(1,255))+'.'+str(r.randint(1,255))
    else:
        str1=''
        str2=''
        for j in range(4):
            p1=r.randint(1,255)
            p2=r.randint(0,255-p1)
            str1+=str(p1)
            str2+=str(p1+p2)
            if(j<3):
                str1+='.'
                str2+='.'
        rules[i][3]=str1+'-'+str2
            
print (rules)
```

```python
f=open('rules.csv','w')
for i in range (5000):
    f.write(rules[i][0]+','+rules[i][1]+','+rules[i][2]+','+rules[i][3])
    if(i!=5000):
        f.write('\n')
```
